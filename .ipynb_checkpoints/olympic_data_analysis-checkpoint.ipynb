```python
# getting the count of number of gold medals by counrty 
df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```

```python
# there is an issue with our data : medals couted with respect to the player ,
# and if one team winning a medal  its showing count for all the member of team 
```

```python
# now the question is how will we solve this isssue?
# dropping the duplicates where team noc games soprts are same
```

```python
medal_tally=df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally=medal_tally.groupby('Region').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```

```python
medal_tally[medal_tally['NOC']=='IND']
```

```python
medal_tally["total"]=medal_tally["Gold"]+medal_tally["Bronze"]+medal_tally["Silver"]
```
